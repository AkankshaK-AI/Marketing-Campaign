```python
# Checking % of outlier values in each numeric variable. 
# Considering a value to be an outlier if it falls beyond the range [mean-2*std, mean+2*std]

bank.describe()
```

```python
import numpy as np

def outlier(df):
    for i in df.describe().columns:
        mean=df.describe().at["mean",i]
        std=df.describe().at["std",i]
        lower_range=mean-(2*std)
        upper_range=mean+(2*std)
        
        x=np.array(df[i])
        p=[]
        for j in x:
            if j<lower_range or j>upper_range:
                p.append(1)
            else:
                p.append(0)
                
        if i=="age":
            df["age_out"]=p
        elif i=="balance":
            df["balance_out"]=p
        elif i=="day":
            df["day_out"]=p
        elif i=="duration":
            df["duration_out"]=p
        elif i=="campaign":
            df["camp_out"]=p
        elif i=="pdays":
            df["pdays_out"]=p
        else:
            df["prev_out"]=p
     
    return df

outlier(bank)
                           
                           
```

```python
# checking % outliers of "previous" variable

bank["prev_out"].mean()
```

```python
# Creating a dataframe to show % outlier for each variable

final=pd.DataFrame()

l=["age_out","balance_out","day_out","duration_out","camp_out","pdays_out","prev_out"]

for i in l:
    new=pd.DataFrame({"variable":i,
                     "%_outlier":[bank[i].mean()]})
    
    final=final.append(new)

final
```
